```python
from bs4 import BeautifulSoup
import requests
import csv
import os
import pandas as pd
import numpy as np
import json
import re
import ast
import datetime
import time
import pickle
```

```python
timestamp = '20230709'
```

```python
stock_df = pd.read_csv(f'{timestamp}_stocks.csv')
```

```python
# Needed to prevent 403 error
header = {'User-Agent': 'Mozilla/5.0'}
```

```python
# Simple way fof tracking if I already scraped the stock
try:
    with open(f'{timestamp}_progress_list.pkl', 'rb') as f:
        progress_list = pickle.load(f)
except:
    progress_list = []
```

```python
print(progress_list)
```

```python
# I was using it with proxy
i = 0
for symbol in stock_df['symbol']:
    if symbol not in progress_list:
        print(f'Downloading {symbol}')
        progress_list.append(symbol)
        
        page = requests.get(url=f'https://www.zacks.com/stock/research/{symbol}/earnings-calendar?tab=earnings', headers=header)
        soup = BeautifulSoup(page.content, 'lxml')
        
        with open(f'../work/earnings_raw_data/{timestamp}_{symbol}.html', 'w') as f:
            f.write(str(soup))
        i += 1
        time.sleep(2)
    if i == 8000:
        break
```

```python
with open(f'{timestamp}_progress_list.pkl', 'wb') as f:
     pickle.dump(progress_list, f)
```
